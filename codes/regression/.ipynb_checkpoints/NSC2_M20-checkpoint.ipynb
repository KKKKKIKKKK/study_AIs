```python
import pymongo as mg
import pandas as pd 
```

```python
client = mg.MongoClient(host='mongodb://localhost:27017')
```

```python
database = client['db_NHIS'] #Database 선택
```

```python
collection = database['NSC2_M20']
```

```python
cursor = collection.find({'SICK_SYM1':'I10'}) # 'SICK_SYM1':'I10' 고혈압 환자만 SELECT 
```

```python
list_BM = list(cursor)
list_BM
```

```python
df_BM = pd.DataFrame(list_BM)
df_BM
```

```python
df_BM_HP=df_BM[['EDC_SBA','MDCARE_STRT_DT', 'TOT_PRSC_DD_CNT', ]]
df_BM_HP
```

```python
df_BM_HP.to_csv('../datasets/NSC2_M20.CSV')
```
